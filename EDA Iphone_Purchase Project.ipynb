### out of 198 F , Iphone purchased  by 77
### out of 182 M , Iphone purchased  by 63

```python
#### Gender Wise Iphone Purchased
plt.figure(figsize=(8,6))
sns.countplot(x = df["Gender"],hue = df["Purchase Iphone"])
plt.savefig('Gender Wise Iphone Purchased.jpg',bbox_inches='tight',dpi=200)
plt.title("Gender Wise Iphone purchased")
plt.show()
```

```python
#### Gender and age Wise Iphone Purchased
plt.figure(figsize=(8,6))
sns.barplot(x = df["Gender"],y =df["Age"],hue = df["Purchase Iphone"])
plt.savefig('Gender-age wise Iphone purchased.jpg',bbox_inches='tight',dpi=200)
plt.title("Gender and Age Wise Iphone purchased")
plt.show()
```

```python
#### Gender and salary wise iphone purchased
plt.figure(figsize=(8,6))
sns.barplot(x = df["Gender"], y = df["Salary"],hue = df["Purchase Iphone"],data=df)
plt.savefig('Gender-Salary wise Iphone purchased.jpg',bbox_inches='tight',dpi=200)
plt.title("Gender and salary wise iphone purchased")
plt.show()
```

```python
#### Age and salary wise iphone purchased
plt.figure(figsize=(8,6))
sns.lineplot(x = df["Age"], y = df["Salary"], hue = df["Purchase Iphone"],data=df)
plt.savefig('Age-Salary wise Iphone purchased.jpg',bbox_inches='tight',dpi=200)
plt.title("Age and Salary wise iphone purchased")
plt.show()
```

```python
sns.displot(x = df["Age"],data=df,kde = True)
plt.savefig('Age.jpg',bbox_inches='tight',dpi=200)
plt.show()
```

```python
sns.displot(x = df["Salary"],data=df,kde = True)
plt.savefig('Salary.jpg',bbox_inches='tight',dpi=200)
plt.show()
```

```python
sns.displot(x = df["Purchase Iphone"],data=df,kde = True)
plt.savefig('Purchased Iphone.jpg',bbox_inches='tight',dpi=200)
plt.show()
```

### Checking  the outliers

```python
plt.figure(figsize=(30,15))
plt.subplot(2,2,1)
sns.boxplot(x = "Age",  data = df)
plt.subplot(2,2,2)
sns.boxplot(x = "Salary",  data = df)
plt.subplot(2,2,3)
sns.boxplot(x = "Purchase Iphone",  data = df)
plt.savefig('Outliers.jpg',bbox_inches='tight',dpi=200)
plt.show()
```
